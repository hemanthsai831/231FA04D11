```python
from transformers import pipeline
```

```python
a = pipeline("sentiment-analysis")
```

```python
text = input("Enter the text:")
print(a(text))
```

```python
gen = pipeline('text-generation')
```

```python
text = input("Enter the text:")
t =gen(text, max_length = 200)
```

```python
print(t[0].get('genetared_text'))
```

```python
n = pipeline('ner')
```

```python
text = input("Enter the text:")
result = n(text)
(result)
```

```python
qa = pipeline('question-answering')
```

```python
c = '''Bity bought some butter,but the butter was bitter butter.
So Bity bought some better butter to make the bitter butter better.'''
```

```python
q = "What does bitty bought?"
```

```python
result = qa({'question': q, 'context': c})
print(result)
```

```python
s = pipeline('summarization')
```

```python
t = '''Bity bought some butter,but the butter was bitter butter.
So Bity bought some better butter to make the bitter butter better.'''
```

```python
result = s(t)
print(result[0].get('summary_text'))
```

```python
u = pipeline('fill-mask')
```

```python
result = u('Bity bought some <mask> butter')
(result)
```
